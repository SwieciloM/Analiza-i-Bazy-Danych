```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Wczytywanie danych gotowych do analizy
df = pd.read_csv("..\Analysis Data\processed_drinks.csv")

# Wyświetlenie kilku pierwszych rekordów  
df.head()
```

```python
# Stworzenie wartości do wykresu
row_sums = df.sum(axis=0)

data = {'Piwo':row_sums.beer_servings, 'Spirytus':row_sums.spirit_servings, 'Wino':row_sums.wine_servings}

courses = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (10, 5))

# Stworzenie wykresu słupkowego
plt.bar(courses, values, color ='maroon', width = 0.4)

plt.xlabel("Rodzaj napoju")
plt.ylabel("Ilość")
plt.title("Łączna liczba porcji napojów na osobę")

# Zapisanie wykresu do pliku
plt.savefig('..\Documents\Charts\char1.jpg')

plt.show()
```
