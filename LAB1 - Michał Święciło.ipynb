## Zadanie 3

Zadanie polegało na zdefiniowaniu funkcji $f(x)=x^2+5$ i sporządzeniu jej wykresu dla argumentów z przedziału:

1. $x>-1$ oraz $x<1$
2. $x>-6$ oraz $x<6$
3. $x>0$ oraz $x<5$

Do wykresów należało dodać etykiety osi, tytuły i legendy.

```python
# Funkcja tworząca wykres funkcji x^2+5 dla wybranego przedziału
def ploter(x1 : float, x2 : float, num_of_points : int = 100):

    # Pierwszy argument większy lub równy drugiemu? 
    if x1 >= x2:
        return None

    # Stworzenie listy argumentów i wartości funkcji
    x = np.linspace(x1, x2, num_of_points)
    y = np.power(x, 2) + 5

    # Rysowanie wykresu i jego elementów
    fig, ax = plt.subplots()
    ax.plot(x, y, label='x^2+5')
    ax.set(xlabel='Argumenty x', ylabel='Wartości y', title='Wykres funkcji dla {}<x<{}'.format(x1, x2))
    ax.legend()
    ax.grid()
    plt.show()
```

```python
# Stworzenie wykresu funkcji dla -1<x<1
ploter(-1, 1)
```

```python
# Stworzenie wykresu funkcji dla -6<x<6
ploter(-6, 6)
```

```python
# Stworzenie wykresu funkcji dla 0<x<5
ploter(0, 5)
```

Aby wykonać to zadanie należało utwórzyć dataframe, w którym kolumny mają nazwy: name, surname, age, sex. Następnie musiał być on uzupełniony pięcioma dowolnymi rekordami. W końcowym etapie powinny zostać wyświetlone informacje o danych (pandas inf), opis danych (pandas describe) oraz pierwsze trzy rekordy (pandas head).

```python
# Tworzenie danych
d = np.array([['Michał', 'Kowalski', 32, 'male'],
              ['Justyna', 'Tomaszewska', 57, 'female'],
              ['Karol', 'Wojtylski', 25, 'male'],
              ['Andrzej', 'Dudniasty', 17, 'male'],
              ['Iwona', 'Robak', 49, 'female']])
# Tworzenie indeksów
i = np.arange(1, len(d)+1, 1)
# Tworzenie nazw kolumn
c = ['NAME', 'SURNAME', 'AGE', 'SEX']
# Tworzenie dataframea
first_df = pd.DataFrame(data=d, index=i, columns=c)
```

```python
# Wyświetlenie zawartości dataframea
print(first_df)
```

```python
# Wyświetlenie informacji związanych z typami danych, wartościami None i zużyciem pamięci wszystkich kolumn
print(first_df.info(verbose=True))
```

```python
# Wyświetlenie statystyk opisowych obejmujących tendencje centralną, rozproszenie i kształt rozkładu zbioru danych
print(first_df.describe())
```
